### Making a bag of words of type of injuries:

Delete NaN entries:

```python
injury_text = data['injury']
```

```python
injury_text = injury_text.dropna()
```

Create array of strings:

```python
injury_text = np.array(injury_text)
```

Delete stop words:

```python
stop_words = stopwords.words('english')
```

```python
injury_text = [i for i in injury_text if i not in stop_words]
```

Create bag of words:

```python
count = CountVectorizer()
bag_of_words = count.fit_transform(injury_text)
bag_of_words.toarray()
```

View BoW headers:

```python
names = count.get_feature_names()
```

BoW and vectors as dataframe:

```python
bow_df = pd.DataFrame(bag_of_words.toarray(),columns=names)
bow_df.head()
```

Drop the columns that are not words:

```python
drop_cols = list(bow_df.columns[:110])
```

```python
bow_df = bow_df.drop(drop_cols, axis =1)
```

```python
bow_df.head()
```

Most frequent words:

```python
bow_df.sum().sort_values(ascending=False).head(10)
```
